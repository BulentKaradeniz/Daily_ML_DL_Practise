```python
import cv2

# Resim dosyasn ykleyin
image = cv2.imread('bulent.jpg')

# Yeni genilik ve ykseklik deerlerini belirleyin
yeni_genislik = 600  # yeni genilik piksel cinsinden
yeni_yukseklik = 600  # Yeni ykseklik piksel cinsinden

# Yeniden boyutlandrma ilemini gerekletirin
yeniden_boyutlandirilmis_resim = cv2.resize(image, (yeni_genislik, yeni_yukseklik))

# Yeniden boyutlandrlm resmi kaydedin
cv2.imwrite('bulent_1.jpg', yeniden_boyutlandirilmis_resim)
```

```python
import cv2
import os  # os modln ie aktarn


# Resim dosyasn ykleyin
image = cv2.imread('derya.jpg')

# Yeni genilik ve ykseklik deerlerini belirleyin
yeni_genislik = 1000  # Minimum kabul edilebilir genilik
yeni_yukseklik = 1000  # Minimum kabul edilebilir ykseklik

# Resmi belirtilen boyutlara yeniden boyutlandrn
yeniden_boyutlandirilmis_resim = cv2.resize(image, (yeni_genislik, yeni_yukseklik))

# Yeniden boyutlandrlm resmi kaydedin
cv2.imwrite('dy1.jpg', yeniden_boyutlandirilmis_resim)

# JPEG formatnda kaydetme
cv2.imwrite('damla2.jpg', yeniden_boyutlandirilmis_resim, [int(cv2.IMWRITE_JPEG_QUALITY), 90])  # Kalite ayarn istediiniz gibi deitirebilirsiniz

# Dosya boyutunu kontrol edin ve gerekiyorsa sktrn
dosya_boyutu = os.path.getsize('damla2.jpg')  # Dosya boyutunu aln
if dosya_boyutu > 240 * 1024:  # 240 KB'den bykse sktrn
    sktrma_oran = 20  # Maksimum 20:1 sktrma
    yeni_kalite = int(90 / sktrma_oran)  # Kaliteyi ayarlayn
    cv2.imwrite('damla2.jpg', yeniden_boyutlandirilmis_resim, [int(cv2.IMWRITE_JPEG_QUALITY), yeni_kalite])  # Yeni kaliteyle kaydet

```

```python
import cv2
import os

# Resim dosyasn ykleyin
image = cv2.imread('dy1.jpg')

# Yeni genilik ve ykseklik deerlerini belirleyin
yeni_genislik = 1200  # Minimum kabul edilebilir genilik
yeni_yukseklik = 1200  # Minimum kabul edilebilir ykseklik

# Resmi belirtilen boyutlara yeniden boyutlandrn
damla3 = cv2.resize(image, (yeni_genislik, yeni_yukseklik))

# Yeniden boyutlandrlm resmi kaydedin
cv2.imwrite('damla3.jpg', damla3)

# JPEG formatnda kaydetme
cv2.imwrite('damla3.jpg', damla3, [cv2.IMWRITE_JPEG_QUALITY, 90])  # Kalite ayarn istediiniz gibi deitirebilirsiniz

# Dosya boyutunu kontrol edin ve gerekiyorsa sktrn
dosya_boyutu = os.path.getsize('damla3.jpg')  # Dosya boyutunu aln
if dosya_boyutu > 240 * 1024:  # 240 KB'den bykse sktrn
    sktrma_oran = 15  # Maksimum 20:1 sktrma
    yeni_kalite = int(90 / sktrma_oran)  # Kaliteyi ayarlayn
    cv2.imwrite('damla3.jpg', damla3, [cv2.IMWRITE_JPEG_QUALITY, yeni_kalite])  # Yeni kaliteyle kaydet
```
