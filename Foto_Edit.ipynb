```python
import cv2

# Resim dosyasını yükleyin
image = cv2.imread('bulent.jpg')

# Yeni genişlik ve yükseklik değerlerini belirleyin
yeni_genislik = 600  # yeni genişlik piksel cinsinden
yeni_yukseklik = 600  # Yeni yükseklik piksel cinsinden

# Yeniden boyutlandırma işlemini gerçekleştirin
yeniden_boyutlandirilmis_resim = cv2.resize(image, (yeni_genislik, yeni_yukseklik))

# Yeniden boyutlandırılmış resmi kaydedin
cv2.imwrite('bulent_1.jpg', yeniden_boyutlandirilmis_resim)
```

```python
import cv2
import os  # os modülünü içe aktarın


# Resim dosyasını yükleyin
image = cv2.imread('derya.jpg')

# Yeni genişlik ve yükseklik değerlerini belirleyin
yeni_genislik = 1000  # Minimum kabul edilebilir genişlik
yeni_yukseklik = 1000  # Minimum kabul edilebilir yükseklik

# Resmi belirtilen boyutlara yeniden boyutlandırın
yeniden_boyutlandirilmis_resim = cv2.resize(image, (yeni_genislik, yeni_yukseklik))

# Yeniden boyutlandırılmış resmi kaydedin
cv2.imwrite('dy1.jpg', yeniden_boyutlandirilmis_resim)

# JPEG formatında kaydetme
cv2.imwrite('damla2.jpg', yeniden_boyutlandirilmis_resim, [int(cv2.IMWRITE_JPEG_QUALITY), 90])  # Kalite ayarını istediğiniz gibi değiştirebilirsiniz

# Dosya boyutunu kontrol edin ve gerekiyorsa sıkıştırın
dosya_boyutu = os.path.getsize('damla2.jpg')  # Dosya boyutunu alın
if dosya_boyutu > 240 * 1024:  # 240 KB'den büyükse sıkıştırın
    sıkıştırma_oranı = 20  # Maksimum 20:1 sıkıştırma
    yeni_kalite = int(90 / sıkıştırma_oranı)  # Kaliteyi ayarlayın
    cv2.imwrite('damla2.jpg', yeniden_boyutlandirilmis_resim, [int(cv2.IMWRITE_JPEG_QUALITY), yeni_kalite])  # Yeni kaliteyle kaydet

```

```python
import cv2
import os

# Resim dosyasını yükleyin
image = cv2.imread('dy1.jpg')

# Yeni genişlik ve yükseklik değerlerini belirleyin
yeni_genislik = 1200  # Minimum kabul edilebilir genişlik
yeni_yukseklik = 1200  # Minimum kabul edilebilir yükseklik

# Resmi belirtilen boyutlara yeniden boyutlandırın
damla3 = cv2.resize(image, (yeni_genislik, yeni_yukseklik))

# Yeniden boyutlandırılmış resmi kaydedin
cv2.imwrite('damla3.jpg', damla3)

# JPEG formatında kaydetme
cv2.imwrite('damla3.jpg', damla3, [cv2.IMWRITE_JPEG_QUALITY, 90])  # Kalite ayarını istediğiniz gibi değiştirebilirsiniz

# Dosya boyutunu kontrol edin ve gerekiyorsa sıkıştırın
dosya_boyutu = os.path.getsize('damla3.jpg')  # Dosya boyutunu alın
if dosya_boyutu > 240 * 1024:  # 240 KB'den büyükse sıkıştırın
    sıkıştırma_oranı = 15  # Maksimum 20:1 sıkıştırma
    yeni_kalite = int(90 / sıkıştırma_oranı)  # Kaliteyi ayarlayın
    cv2.imwrite('damla3.jpg', damla3, [cv2.IMWRITE_JPEG_QUALITY, yeni_kalite])  # Yeni kaliteyle kaydet
```
